```python
import os
import warnings

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/S&P500_monthly_full.csv")
```

```python
df
```

```python
df["DATE"] = pd.to_datetime(df["DATE"])
```

```python
df = df.sort_values("DATE").reset_index(drop=True)
```

```python
df = df.dropna(axis=1)
```

```python
df["SP500_full"] = df["SP500_full"].str.replace("\? ", "")
```

```python
df["SP500_full"] = df["SP500_full"].str.replace(",", "")
```

```python
df
```

```python
df.to_csv("../data/raw/SP500_monthly_full.csv", index=False)
```
